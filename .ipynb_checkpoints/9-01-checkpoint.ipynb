```python
import random
rand_num = 1
while rand_num != 4:
    rand_num = random.randint(1,6)
    print(rand_num)
```

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num) 
    if rand_num == 4:
        break
    print('.')
```

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
    print('.')
```

```python
for x in range(1, 100):
    if x % 3 == 0:
        if x % 5 == 0:
            print('Fizz Bazz')
    else:
        print(x)
        
    
```

```python
for x in range(1, 100):
        if x % 3 == 0:
            if x % 5 == 0:
                print('Fizz Bazz')
            else:
                print('Fizz')
        elif x % 5 == 0:
            print('Bazz')
        else:
            print(x)
        
    
```
